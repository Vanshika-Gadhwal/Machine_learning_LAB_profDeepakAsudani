```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X= np.array([1., 2., 3., 4., 5.])
y= np.array([0., 0., 0., 1., 1.,])
m=len(y)
```

```python
def sigmoid(z):
     return 1.0 / (1.0 + np.exp(-z))
```

```python
b0 = 0.0
b1=0.0
alpha=0.1
```

```python
z=b0+b1*X
h = sigmoid(z)
eps = 1e-15
initial_cost = - (1.0 / m) * np.sum(y * np.log(h + eps) + (1 - y) * np.log(1 - h + eps))
print("Initial h:", h)
print("Initial cost:", initial_cost)
```

```python
h_minus_y = h - y
grad_b0 = np.mean(h_minus_y)
grad_b1 = np.mean(h_minus_y * X)
print("grad_b0 =", grad_b0)
print("grad_b1 =", grad_b1)
```

```python
b0_new = b0 - alpha * grad_b0
b1_new = b1 - alpha * grad_b1
print("b0_new =", b0_new)
print("b1_new =", b1_new)
```

```python

b0, b1 = 0.0, 0.0
n_iter = 2000
history = []
for it in range(n_iter):
    z = b0 + b1 * X
    h = sigmoid(z)
    grad_b0 = np.mean(h - y)
    grad_b1 = np.mean((h - y) * X)
    cost = - (1.0 / m) * np.sum(y * np.log(h + eps) + (1 - y) * np.log(1 - h + eps))
    history.append(cost)
    b0 -= alpha * grad_b0
    b1 -= alpha * grad_b1
```

```python
# One gradient descent update
b0_new = b0 - alpha * grad_b0
b1_new = b1 - alpha * grad_b1
print("After one update: b0 =", b0_new, ", b1 =", b1_new)

# Run gradient descent
b0, b1 = 0.0, 0.0
n_iter = 2000
history = []

for it in range(n_iter):
    z = b0 + b1 * X
    h = sigmoid(z)

    grad_b0 = np.mean(h - y)
    grad_b1 = np.mean((h - y) * X)
    cost = -(1.0 / m) * np.sum(y * np.log(h + eps) + (1 - y)
                                * np.log(1 - h + eps))

    history.append((it, b0, b1, cost))

    b0 = b0 - alpha * grad_b0
    b1 = b1 - alpha * grad_b1

print("\nFinal parameters after {} iterations:".format(n_iter))
print("b0 =", b0)
print("b1 =", b1)

```

```python
print("\nFinal parameters after {} iterations:".format(n_iter))
print("b0 =", b0)
print("b1 =", b1)
decision_boundary = -b0 / b1
print("Decision boundary x =", decision_boundary)
print("Final cost J =", history[-1][3])
```

```python
X_plot = np.linspace(0, 6, 200)
y_plot = sigmoid(b0 + b1 * X_plot)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y)
plt.plot(X_plot, y_plot)
plt.axhline(0.5, linestyle='--')
plt.title("Logistic regression fit (manual gradient descent)")
plt.xlabel("Study hours")
plt.ylabel("P(pass)")
plt.grid(True)
plt.show()
```
